```python
#Which 5 year Period India saw huge win percent?
```

```python
# Creating a column for the year a match was played
df2["Year"] = df2["Date"].apply(lambda x: int(x.split("/")[-1]))
```

```python
df2.head()
```

```python
win_ratio = []
i = 0
years = []
while True:
    start = df2["Year"][i]
    end = start+5
    try:
        df = df2[(df2["Year"]>=start) & (df2["Year"]<=end)]["Verdict"]
    except Exception as e:
        df = df2[(df2["Year"]>=start) & (df2["Year"]<=2021)]["Verdict"] 
    win_ratio.append(df.sum()/len(df))
    years.append(end)
    if i >= len(df2):
        break
    #print(df)
    i = i+len(df)
    #print(len(df2))
    if i >= len(df2):
        break
    
```

```python
win_ratio
```

```python
plt.plot(years,win_ratio)
plt.title("WinRatio over 5 year time periods")
plt.xlabel("Year")
plt.ylabel("Win_ration")
```

We can see that win ratio for India has been increasing
